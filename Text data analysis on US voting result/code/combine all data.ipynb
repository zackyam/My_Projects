```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.metrics import classification_report
from sklearn.model_selection import train_test_split
from sklearn import metrics
from sklearn.model_selection import GridSearchCV
from sklearn.linear_model import LinearRegression, Ridge, Lasso
from sklearn.neighbors import KNeighborsClassifier
from sklearn.linear_model import LogisticRegression
from sklearn import tree
from sklearn.svm import SVC
import statsmodels.api as sm
from sklearn import preprocessing
from sklearn.tree import DecisionTreeClassifier, plot_tree
```

```python
election_result = pd.read_csv("election_result.csv")
```

```python
election_result = election_result.rename(columns={'county_fips': 'FIPS Code'})
```

```python
election_result
```

```python
Education = pd.read_csv("Education.csv")
```

```python
Education = Education.drop(["State","Area name"], axis=1)
```

```python
PopulationEstimates = pd.read_csv("PopulationEstimates.csv")
```

```python
PopulationEstimates = PopulationEstimates.drop(["State","Area_Name"], axis=1)
```

```python
PopulationEstimates = PopulationEstimates.rename(columns={'FIPStxt': 'FIPS Code'})
```

```python
PovertyEstimates = pd.read_csv("PovertyEstimates.csv")
```

```python
PovertyEstimates = PovertyEstimates.drop(["Stabr","Area_name"], axis=1)
PovertyEstimates = PovertyEstimates.rename(columns={'FIPStxt': 'FIPS Code'})
```

```python
Unemployment = pd.read_csv("Unemployment.csv")
```

```python

Unemployment = Unemployment.drop(["Stabr","area_name"], axis=1)
Unemployment = Unemployment.rename(columns={'fips_txt': 'FIPS Code'})
```

```python
Unemployment
```

```python
df = Education.merge(PopulationEstimates, on='FIPS Code', how='inner', suffixes=('_1', '_2'))
```

```python
df = df.merge(PovertyEstimates, on='FIPS Code', how='inner', suffixes=('_1', '_2'))
```

```python
df = df.merge(Unemployment, on='FIPS Code', how='inner', suffixes=('_1', '_2'))
```

```python
df
```

```python
final_df = election_result.merge(df, on='FIPS Code', how='inner', suffixes=('_1', '_2'))
```

```python
final_df
```

```python
final_df.to_csv("final_df.csv",index=False)
```
