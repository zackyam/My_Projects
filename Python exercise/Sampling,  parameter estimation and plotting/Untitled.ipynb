```python
import numpy as np
```

```python
data = open("data.txt","r")
```

```python
lines = data.readlines()
dataused = lines[3:]
```

```python
new_data = [x[:-1] for x in dataused]
```

```python
arr = np.array(new_data)
```

```python

data = arr.astype(np.float)
```

```python
n = 5
sample = [data[k:k+n] for k in range(0, len(new_data), n)]
```

```python
sample_mean = ser.groupby(ser.index//5).mean()
```

```python
def create_bins(n, delta):
    quantity = 100000/n

    bins = []
    for low in np.arange(0.0, 0.0+ quantity*delta, delta):
        bins.append((low, low+delta))
    return bins
```
