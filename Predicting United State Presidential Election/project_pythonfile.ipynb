```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.metrics import classification_report
from sklearn.model_selection import train_test_split
from sklearn import metrics
from sklearn.model_selection import GridSearchCV
from sklearn.linear_model import LinearRegression, Ridge, Lasso
from sklearn.neighbors import KNeighborsClassifier
from sklearn.linear_model import LogisticRegression
from sklearn import tree
from sklearn.svm import SVC
import statsmodels.api as sm
from sklearn import preprocessing
from sklearn.tree import DecisionTreeClassifier, plot_tree
```

## Data Cleaning

```python
countywithnames = pd.read_csv("county_facts.csv")
```

```python
countywithnames.dropna(inplace=True)
```

```python
county_facts = countywithnames.set_index("fips")
```

```python
county_facts
```

```python
facts_used = county_facts.drop(['area_name', 'state_abbreviation'], axis = 1) 
```

```python
facts_used
```

```python
votes = pd.read_csv("votes.csv")
```

```python
votes_2016 = votes.set_index("fips")
```

```python
def dummyvote(row):
    if row['Trump'] > row['Clinton']:
        val = 1
    else:
        val = 0
    return val


votes_2016['Rep2016'] = votes_2016.apply(dummyvote, axis=1)
```

```python
def dummyvotes(row):
    if row['Romney'] > row['Obama']:
        val = 1
    else:
        val = 0
    return val


votes_2016['Rep2012'] = votes_2016.apply(dummyvotes, axis=1)
```

```python
votes_2016
```

```python
vote_result = votes_2016[["Rep2016","Rep2012"]]
```

```python
vote_result
```

```python
merged_data = vote_result.merge(facts_used,how="left",on=["fips"])
```

```python
merged_data = merged_data.drop(["POP010210","PST040210"], axis=1)
```

## Split data

```python
y = merged_data["Rep2016"]
X = merged_data.drop(['Rep2016'], axis = 1) 
```

```python
x_in, x_out, y_in, y_out = train_test_split(X, y, test_size=0.25)
```

```python
x_in
```

```python
x_scaled = preprocessing.scale(x_in)
```

```python
x_out_scaled = preprocessing.scale(x_out)
```

```python
target_names =  ["Rep Lose","Rep Win"]
```

## Predict voting result:

## Logist

```python
grid={ 'C': [0.001,0.01,0.1,1,10,100,1000], "penalty":["l2"], "max_iter":[4000]}
logist=LogisticRegression()
logist_cv=GridSearchCV(logist,grid,cv=10)
logist_cv.fit(x_scaled,y_in)

print("tuned hpyerparameters :(best parameters) ",logist_cv.best_params_)
print("accuracy :",logist_cv.best_score_)
```

```python
y_logit_pred = logist_cv.predict(x_out_scaled)
print(classification_report(y_out, y_logit_pred, target_names=target_names))
```

## Nearest Neighbor 

```python
grid={ 'n_neighbors': [1,3,7,9,11,13,15], "weights":["uniform","distance"], "metric":["euclidean","manhattan"]}
knn = KNeighborsClassifier()
knn_cv = GridSearchCV(knn,grid,cv=10)
knn_cv.fit(x_scaled,y_in)

print("tuned hpyerparameters :(best parameters) ",knn_cv.best_params_)
print("accuracy :",knn_cv.best_score_)
```

```python
y_KNN_pred = knn_cv.predict(x_out_scaled)
print(classification_report(y_out, y_KNN_pred, target_names=target_names))
```

## Decision Tree

```python
grid = {"criterion" : ["gini","entropy"] , "max_depth":[2,4,6,8,10] , "min_samples_split":[2,4,6,8,10] , "min_samples_leaf":[2,4,6,8,10]}
dt = tree.DecisionTreeClassifier()
dt_cv = GridSearchCV(dt,grid,cv=10)
dt_cv.fit(x_scaled,y_in)

print("tuned hpyerparameters :(best parameters) ",dt_cv.best_params_)
print("accuracy :",dt_cv.best_score_)
```

```python
y_DT_pred = dt_cv.predict(x_out_scaled)
print(classification_report(y_out, y_DT_pred, target_names=target_names))
```
